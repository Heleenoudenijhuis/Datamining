```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 15, 15
```

```python
df = (pd.read_csv('training_set_VU_DM.csv'))
print('loading the dataset, make sure it is in the same folder')
```

```python
#datetime features
df['date_time'] =  pd.to_datetime(df['date_time'])
df['day']=df['date_time'].dt.day
df['month']= df['date_time'].dt.month
df['year']=df['date_time'].dt.year

#Find booking date
df['date_time_book'] = df['date_time'] -  pd.to_timedelta(df['srch_booking_window'], unit='d')
df.head()
```

```python
click_data = df.loc[df['click_bool'] == 1]
click_data.shape
df = df.loc[df['click_bool'] == 0]
```

```python
unique = df.srch_id.unique()

#random filter of 10,000 users
idf = random.sample(list(unique), 10000)
dff = df[df.srch_id.isin(idf)]

dff = click_data.append(dff)
```

```python
dff.to_csv('filtered_training_set.csv')
```

```python
df = (pd.read_csv('filtered_training_set.csv'))
df = df.drop(df.columns[0], axis=1)
df
```

```python
#divide into 80/20 train/test set

train = df.iloc[0:(len(df)//10*8),]
test = df.iloc[(len(df)//10*8):len(df),]

#remove 4 columns (position, click_bool, booking_bool and gross_bookings_usd) 
#as they are also not in test_set_VU_DM.csv
#test2 = test.drop(columns=['position', 'click_bool', 'booking_bool', 'gross_bookings_usd'])

#export to csv files
train.to_csv('train.csv', index=False)
test.to_csv('test.csv', index=False)
#test2.to_csv('test2.csv', index=False)
```
