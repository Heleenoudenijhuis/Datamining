```python
# Importing packages and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the data
df = pd.read_csv('ODI-2020.csv', sep=';')
df
```

```python
# Analysis of the first column: What programme are you in
df['What programme are you in?'] = [x.lower() for x in df['What programme are you in?']]
program = df['What programme are you in?'].unique().tolist()
program

# Question is how we are gonna clean this data
```

```python
# Analysis of the second column: Have you taken a course on machine learning?
machine = df['Have you taken a course on machine learning?'].value_counts()
machine = pd.DataFrame(machine)
plot = machine.plot.pie(y='Have you taken a course on machine learning?', figsize=(7, 7))
```

```python
# Analysis of the third column: Have you taken a course on information retrieval?
info = df['Have you taken a course on information retrieval?'].value_counts()
info = pd.DataFrame(info)
plot = info.plot.pie(y='Have you taken a course on information retrieval?', figsize=(7, 7))
```

```python
# Analysis of the fourth column: Have you taken a course on statistics?
stat = df['Have you taken a course on statistics?'].value_counts()
stat = pd.DataFrame(stat)
plot = stat.plot.pie(y='Have you taken a course on statistics?', figsize=(7, 7))
```

```python
# Analysis of the fifth column: Have you taken a course on Have you taken a course on databases?
data = df['Have you taken a course on databases?'].value_counts()
data = pd.DataFrame(data)
plot = data.plot.pie(y='Have you taken a course on databases?', figsize=(7, 7))
```

```python
# Analysis of the sixth column: What is your gender?
gender = df['What is your gender?'].value_counts()
gender = pd.DataFrame(gender)
plot = gender.plot.pie(y='What is your gender?', figsize=(7, 7))
```

```python
# Analysis of the seventh column: Chocolate makes you.....
choc = df['Chocolate makes you.....'].value_counts()
choc = pd.DataFrame(choc)
plot = choc.plot.pie(y='Chocolate makes you.....', figsize=(7, 7))
```

```python
# Analysis of the eigth column: When is your birthday (date)?
birth = df['When is your birthday (date)?']
birth

# Data has to be cleaned as well
```

```python
# Analysis of the nineth column: Number of neighbors sitting around you?
neigh = df['Number of neighbors sitting around you?'].value_counts()
neigh = pd.DataFrame(neigh)
plt.bar(neigh.index, neigh['Number of neighbors sitting around you?'])
plt.xticks(rotation=90)
plt.title('Number of neighbors sitting around you?')
plt.ylabel('Count')
plt.show()
```

```python
# Analysis of the tenth column: Did you stand up?
stand = df['Did you stand up?'].value_counts()
stand = pd.DataFrame(stand)
plot = stand.plot.pie(y='Did you stand up?', figsize=(7, 7))
```

```python
# Analysis of the eleventh column: What is your stress level (0-100)?
df = df[~df['What is your stress level (0-100)?'].str.contains("-")]
df['What is your stress level (0-100)? 2'] = pd.to_numeric(df['What is your stress level (0-100)?'],errors='coerce')
drop = df[ (df['What is your stress level (0-100)? 2'] <= 0) & 
                                              (df['What is your stress level (0-100)? 2'] >= 100) ].index
df.drop(drop, inplace=True)
plt.rcParams["figure.figsize"] = (10,5)
plt.hist(df['What is your stress level (0-100)? 2'], bins=100, color='grey')
plt.title('What is your stress level (0-100)?')
plt.ylabel('Count')
plt.xlabel('Stress level')
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python
# Analysis of the twelveth column: Money to be given
df[df.columns[11]]= pd.to_numeric(df[df.columns[11]],errors='coerce')
plt.rcParams["figure.figsize"] = (10,5)
plt.hist(df[df.columns[11]], bins=100, color='grey')
plt.title('DM competition')
plt.ylabel('Count')
plt.xlabel('Amount of money')
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python
# Analysis of the twelveth column: Money to be given
df = df[df[df.columns[11]] < 100] 
plt.rcParams["figure.figsize"] = (10,5)
plt.hist(df[df.columns[11]], bins=100, color='grey')
plt.title('DM competition')
plt.ylabel('Count')
plt.xlabel('Amount of money')
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python
# Analysis of the twelveth column: Give a random number
df = df[~df['Give a random number'].str.contains("-")]
df['Give a random number 2'] = pd.to_numeric(df['Give a random number'],errors='coerce')
drop = df[ (df['Give a random number 2'] <= 0) & 
                                              (df['Give a random number 2'] >= 100) ].index
df.drop(drop, inplace=True)
plt.rcParams["figure.figsize"] = (10,5)
plt.hist(df['Give a random number 2'], bins=100, color='grey')
plt.title('Give a random number')
plt.ylabel('Count')
plt.xlabel('Random number')
plt.xticks(rotation=90, fontsize=10)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
text = df['What makes a good day for you (1)?'].values 
text2 = df['What makes a good day for you (2)?'].values 
joined = [text, text2]
good = np.concatenate((text, text2))

wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(good))
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()

wordcloud = WordCloud().generate(str(good))

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
