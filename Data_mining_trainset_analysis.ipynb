```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train.csv')
```

```python
train
```

```python
# Analysis of sex
sex = train['Sex'].value_counts()
sex = pd.DataFrame(sex)
plot = sex.plot.pie(y='Sex', figsize=(7, 7))
```

```python
# Analysis of survivors and diseased
survived = train[train['Survived'] == 1]
diseased = train[train['Survived'] == 0]

# Sex distribution
sex_s = survived['Sex'].value_counts()
sex_s = pd.DataFrame(sex_s)
plot = sex_s.plot.pie(y='Sex', figsize=(7, 7))
sex_d = diseased['Sex'].value_counts()
sex_d = pd.DataFrame(sex_d)
plot = sex_d.plot.pie(y='Sex', figsize=(7, 7))
```

```python
# Class distribution
class_s = survived['Pclass'].value_counts()
class_s = pd.DataFrame(class_s)
plot = class_s.plot.pie(y='Pclass', figsize=(7, 7))
class_d = diseased['Pclass'].value_counts()
class_d = pd.DataFrame(class_d)
plot = class_d.plot.pie(y='Pclass', figsize=(7, 7))
```

```python
# Age distribution
plt.hist(survived['Age'], bins=50, color='grey')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age distribution amongst survivors')
plt.show()

plt.hist(diseased['Age'], bins=50, color='grey')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age distribution amongst diseased')
plt.show()
```

```python
# Fare distribution
plt.hist(survived['Fare'], bins=100, color='grey')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.xlim(0,100)
plt.title('Fare distribution amongst survivors')
plt.show()

plt.hist(diseased['Fare'], bins=50, color='grey')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.xlim(0,100)
plt.title('Fare distribution amongst diseased')
plt.show()
```

```python
# Sibling distribution
sib_s = survived['SibSp'].value_counts()
sib_s = pd.DataFrame(sib_s)
plot = sib_s.plot.pie(y='SibSp', figsize=(7, 7))
sib_d = diseased['SibSp'].value_counts()
sib_d = pd.DataFrame(sib_d)
plot = sib_d.plot.pie(y='SibSp', figsize=(7, 7))
```

```python
# Embarked distribution
emb_s = survived['Embarked'].value_counts()
emb_s = pd.DataFrame(emb_s)
plot = emb_s.plot.pie(y='Embarked', figsize=(7, 7))
emb_d = diseased['Embarked'].value_counts()
emb_d = pd.DataFrame(emb_d)
plot = emb_d.plot.pie(y='Embarked', figsize=(7, 7))
```
